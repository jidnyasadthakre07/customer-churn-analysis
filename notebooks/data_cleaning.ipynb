```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/telco_churn.csv')
```

```python
print(df.shape)
```

```python
print("Churn Counts:")
print(df['Churn'].value_counts()) 
```

```python
print(df.head())
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
#Fix TotalCharges (it's text, should be number)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
#Check how many became blank after fixing
print("Blank TotalCharges:", df['TotalCharges'].isnull().sum())
```

```python
#Remove customerID (not useful)
df.drop('customerID', axis=1, inplace=True)
```

```python
#Convert Churn column from Yes/No to 1/0
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
print("Churn after encoding:")
print(df['Churn'].value_counts())
```

```python
#Save cleaned data
df.to_csv('../data/telco_churn_cleaned.csv', index=False)
print("✅ Cleaned file saved!")
```
