```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/telco_churn_cleaned.csv')
```

```python
#INSIGHT 1: Overall Churn Rate
churn_rate = df['Churn'].value_counts(normalize=True) * 100
print(churn_rate)
```

```python
sns.countplot(x='Churn', data=df, hue='Churn', palette='Set2', legend=False)
plt.title('Overall Churn Distribution')
plt.xticks([0,1], ['Retained (73.5%)', 'Churned (26.5%)'])
plt.savefig('../visuals/churn_distribution.png')
plt.show()
```

```python
# INSIGHT 2: Contract Type vs Churn 
sns.countplot(x='Contract', hue='Churn', data=df, palette='Set2')
plt.title('Churn by Contract Type')
plt.savefig('../visuals/churn_by_contract.png')
plt.show()
```

```python
# INSIGHT 3: Tenure vs Churn 
sns.boxplot(x='Churn', y='tenure', data=df, hue='Churn', palette='Set2', legend=False)
plt.title('Customer Tenure vs Churn')
plt.xticks([0,1], ['Retained', 'Churned'])
plt.savefig('../visuals/tenure_vs_churn.png')
plt.show()
```

```python
#  INSIGHT 4: Monthly Charges vs Churn ----
sns.boxplot(x='Churn', y='MonthlyCharges', data=df,
            hue='Churn', palette='Set2', legend=False)
plt.title('Monthly Charges vs Churn')
plt.xticks([0,1], ['Retained', 'Churned'])
plt.savefig('../visuals/charges_vs_churn.png')
plt.show()
```
